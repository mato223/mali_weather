## Choix du modèle 

La prédiction des données météorologiques est une tâche complexe en raison de la nature dynamique et non linéaire de l'atmosphère. Les modèles utilisés pour la prédiction météorologique peuvent varier en fonction de la portée de la prédiction (à court terme, à moyen terme ou à long terme) et des caractéristiques spécifiques des données disponibles. Voici quelques-uns des modèles couramment utilisés pour la prédiction météorologique :

1. **Modèles basés sur la physique :**
   - **Modèles de prévision numérique du temps (NWP) :** Ces modèles utilisent des équations physiques pour simuler le comportement de l'atmosphère. Des exemples incluent le modèle WRF (Weather Research and Forecasting) et le modèle ECMWF (European Centre for Medium-Range Weather Forecasts).

2. **Modèles statistiques :**
   - **Modèles de régression :** Les modèles de régression, tels que la régression linéaire, peuvent être utilisés pour modéliser la relation entre différentes variables météorologiques.
   - **Modèles ARIMA (AutoRegressive Integrated Moving Average) :** Ces modèles sont utilisés pour modéliser les séries temporelles et peuvent être adaptés à la prédiction de variables météorologiques.

3. **Réseaux de neurones :**
   - **Réseaux de neurones convolutionnels (CNN) :** Ils peuvent être utilisés pour l'analyse d'images satellite et la prévision météorologique basée sur des images.
   - **Réseaux de neurones récurrents (RNN) et Long Short-Term Memory (LSTM) :** Ils sont adaptés à la modélisation de séquences temporelles et peuvent être utilisés pour la prédiction météorologique.

4. **Ensembles d'arbres de décision :**
   - **Random Forests :** Les forêts aléatoires peuvent être utilisées pour la prédiction météorologique en agrégeant les prédictions de plusieurs arbres de décision.
   - **Gradient Boosted Trees :** Les modèles basés sur le gradient boosting, tels que XGBoost et LightGBM, sont également populaires pour la prédiction météorologique.

5. **Méthodes hybrides :**
   - **Combinaison de modèles physiques et statistiques :** Parfois, les modèles basés sur la physique sont combinés avec des modèles statistiques pour tirer parti de la précision des modèles physiques et de la capacité des modèles statistiques à capturer des modèles plus fins.

Il est important de noter que le choix du modèle dépend du contexte spécifique, de la disponibilité des données, de la portée temporelle de la prédiction et d'autres considérations. Les modèles NWP sont souvent utilisés pour les prévisions à court et moyen terme, tandis que les modèles statistiques et les réseaux de neurones peuvent être utilisés pour des prévisions à plus long terme ou pour des applications spécifiques. Le domaine de la recherche en prédiction météorologique est en constante évolution, et de nouvelles approches et modèles sont régulièrement proposés.

```python
# Sélection des caractéristiques et de la cible
features = data.drop(['temperature_2m', 'date'], axis=1)  # Excluez 'date' car elle est maintenant redondante
target = data['temperature_2m']

# Division des données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)

# Normalisation des données (facultatif, mais souvent recommandé)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Création et entraînement du modèle
model = LinearRegression()
model.fit(X_train_scaled, y_train)

# Prédictions sur l'ensemble de test
predictions = model.predict(X_test_scaled)

# Évaluation du modèle
mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')
```

```python
# Vérifier la stationnarité (Dickey-Fuller Test)

result = adfuller(df1['temperature_2m'])
print('ADF Statistic:', result[0])
print('p-value:', result[1])
print('Critical Values:', result[4])

# Si la série n'est pas stationnaire, appliquer la différenciation pour rendre la série stationnaire
differenced_series = df1['temperature_2m'].diff().dropna()

# Visualisation de la série temporelle différenciée
plt.plot(differenced_series)
plt.title('Différenciation de la série temporelle')
plt.show()

# Diviser les données en ensembles d'entraînement et de test
train_size = int(len(differenced_series) * 0.8)
train, test = differenced_series[0:train_size], differenced_series[train_size:]

# Entraînement du modèle ARIMA
order = (1, 1, 1)  # Les valeurs de p, d, q (ordre AR, ordre de différenciation, ordre MA)
model = ARIMA(train, order=order)
model_fit = model.fit()

# Prédictions sur l'ensemble de test
predictions = model_fit.forecast(steps=len(test))

# Inverse de la différenciation pour obtenir les prédictions originales
predictions_inverse = differenced_series[train_size-1] + predictions.cumsum()

# Évaluation du modèle
mse = mean_squared_error(test, predictions)
rmse = sqrt(mse)

print(f'Mean Squared Error: {mse}')
print(f'Root Mean Squared Error: {rmse}')

# Visualisation des résultats
plt.plot(df1['temperature_2m'], label='Observations réelles')
plt.plot(predictions_inverse, color='red', label='Prédictions ARIMA')
plt.title('Prédictions ARIMA')
plt.legend()
plt.show()
```
